```python
def compare_columns(df1_name, df2_name, df1, df2):
    columns_df1 = set(df1.columns)
    columns_df2 = set(df2.columns)
    
    columns_only_in_df1 = columns_df1 - columns_df2
    columns_only_in_df2 = columns_df2 - columns_df1
    
    if columns_only_in_df1:
        print(f"Columns only in {df1_name}:", columns_only_in_df1)
    else:
        print(f"No columns only in {df1_name}")
    
    if columns_only_in_df2:
        print(f"Columns only in {df2_name}:", columns_only_in_df2)
    else:
        print(f"No columns only in {df2_name}")

compare_columns("original_df", "features_df", original_df, features_df)
```

```python
import pandas as pd

def fill_null_with_median(df, column_name):
    # Convert specified column to numeric, ignoring errors
    df[column_name] = pd.to_numeric(df[column_name])

    # Fill null values in specified column with the median
    df[column_name].fillna(df[column_name].median(), inplace=True)

fill_null_with_median(original_df, 'Distance')
fill_null_with_median(features_df, 'Distance')

```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

def one_hot_encode_column(df1, df2, column_name, new_column_name):
    # Create OneHotEncoder object
    encoder = OneHotEncoder(sparse_output=False)
    
    # Fit and transform the encoder on df1
    encoded_column = encoder.fit_transform(df1[[column_name]])
    
    # Convert the transformed column into a DataFrame
    encoded_df = pd.DataFrame(encoded_column)
    
    # Rename the columns
    encoded_df.columns = [new_column_name + "_" + str(i) for i in range(encoded_df.shape[1])]
    
    # Concatenate the encoded column with df2
    df2 = pd.concat([df2, encoded_df], axis=1)
    
    return df2

```

The square root choice is the default in MS Excel and an early mentioning of this can be found in Davies & Goldsmith 1980, p. 11 (Lohaka, 2007). It sais to take the square root out of the total number of cases, and round this up to an integer. This is then the number of bins that should be used. If you have for example 150 scores, then according to the square root choice method, you should take the square root of 150, which is about 12.25, and then round this up to 13. This 13 is then the number of bins to be used. The lower bound of the first bin should then be the minimum score.

link: (https://peterstatistics.com/Ultimate/Session2/S02E05/S02E05-3-CreatBins/S02E05-3-CreatingBins.html#:~:text=If%20you%20have%20for%20example,then%20be%20the%20minimum%20score.)

```python
import numpy as np

# Calculate the number of data points in the dataset
num_data_points = len(original_df)

# Calculate the number of bins using the Square Root Rule
num_bins = int(np.sqrt(num_data_points))

print("Number of Data Points:", num_data_points)
print("Number of Bins (Square Root Rule):", num_bins)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

def evaluate_model(model, X, y):
    # Make predictions
    predictions = model.predict(X)
    
    # Calculate RMSE
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mse)
    print(f'{type(model).__name__} Root Mean Square Error (RMSE): {rmse:.2f}')
    
    # Calculate MAE
    mae = mean_absolute_error(y, predictions)
    print(f'{type(model).__name__} Mean Absolute Error (MAE): {mae:.2f}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

def plot_learning_curves_and_evaluate(model, X, y, xlim=None, ylim=None, figsize=(7, 4)):
    plt.figure(figsize=figsize)  # Set figure size here
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)
    plt.xlabel("Training set size", fontsize=14)
    plt.ylabel("RMSE", fontsize=14)
    
    # Get the name of the model class
    model_name = type(model).__name__
    
    # Adjust title based on whether axis limits are defined or not
    if xlim is not None and ylim is not None:
        title = f'Validation vs Training in a Specific Range - {model_name}'
    else:
        title = f'Validation vs Training in the whole dataset - {model_name}'
    
    plt.title(title, fontsize=14)
    
    # Apply axis limits if provided
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    
    # Evaluate model
    # Training set
    y_train_predict = model.predict(X_train)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_predict))
    train_mae = mean_absolute_error(y_train, y_train_predict)
    print(f'{model_name} Training Set RMSE: {train_rmse:.2f}')
    print(f'{model_name} Training Set MAE: {train_mae:.2f}')
    
    # Validation set
    y_val_predict = model.predict(X_val)
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_predict))
    val_mae = mean_absolute_error(y_val, y_val_predict)
    print(f'{model_name} Validation Set RMSE: {val_rmse:.2f}')
    print(f'{model_name} Validation Set MAE: {val_mae:.2f}')
    
    plt.show()
```

```python
import random
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Setting random seeds
np.random.seed(20)
tf.random.set_seed(20)
random.seed(20)

# building model
def build_model(clear=True):
    if clear:
        tf.keras.backend.clear_session()

    inputs_layer = tf.keras.layers.Input(shape=(46,), name='input_features')
    inputs_flatten = tf.keras.layers.Flatten()(inputs_layer)

    dense1 = tf.keras.layers.Dense(256, activation='relu')(inputs_flatten)
    dense2 = tf.keras.layers.Dense(128, activation='relu')(dense1)
    dropout1 = tf.keras.layers.Dropout(0.0)(dense2)
    dense3 = tf.keras.layers.Dense(64, activation='relu')(dropout1)
    output_layer = tf.keras.layers.Dense(32, activation='relu')(dense3)

    model = tf.keras.Model(inputs=inputs_layer, outputs=output_layer)

    model.compile(
        optimizer='adam',
        loss='mean_absolute_error',  # Use MAE as loss function
        metrics=[tf.keras.metrics.MeanAbsoluteError()]  # Monitor MAE instead of accuracy
    )

    return model

model = build_model()

# Train
history = model.fit(
    x_train,
    y_train,
    epochs=100,
    batch_size=128,
    validation_data=(x_val, y_val),
    verbose=1,
    callbacks=[tf.keras.callbacks.EarlyStopping(
        monitor='val_mean_absolute_error',  # Monitor MAE on validation data
        patience=30,
        restore_best_weights=True
    )]
)

mae = history.history["mean_absolute_error"]
val_mae = history.history["val_mean_absolute_error"]

epochs = range(1, len(mae) + 1)

# Plotting MAE
plt.plot(epochs, mae, label="Training MAE")
plt.plot(epochs, val_mae, label="Validation MAE")
plt.title("Training and validation MAE for aveOralF")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend()
plt.show()
```
