```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

def one_hot_encode_column(df1, df2, column_name, new_column_name):
    # Create OneHotEncoder object
    encoder = OneHotEncoder(sparse_output=False)
    
    # Fit and transform the encoder on df1
    encoded_column = encoder.fit_transform(df1[[column_name]])
    
    # Convert the transformed column into a DataFrame
    encoded_df = pd.DataFrame(encoded_column)
    
    # Rename the columns
    encoded_df.columns = [new_column_name + "_" + str(i) for i in range(encoded_df.shape[1])]
    
    # Concatenate the encoded column with df2
    df2 = pd.concat([df2, encoded_df], axis=1)
    
    return df2

```

The square root choice is the default in MS Excel and an early mentioning of this can be found in Davies & Goldsmith 1980, p. 11 (Lohaka, 2007). It sais to take the square root out of the total number of cases, and round this up to an integer. This is then the number of bins that should be used. If you have for example 150 scores, then according to the square root choice method, you should take the square root of 150, which is about 12.25, and then round this up to 13. This 13 is then the number of bins to be used. The lower bound of the first bin should then be the minimum score.

link: (https://peterstatistics.com/Ultimate/Session2/S02E05/S02E05-3-CreatBins/S02E05-3-CreatingBins.html#:~:text=If%20you%20have%20for%20example,then%20be%20the%20minimum%20score.)
