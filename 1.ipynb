```python
def compare_columns(df1, df2):
    columns_df1 = set(df1.columns)
    columns_df2 = set(df2.columns)
    
    columns_only_in_df1 = columns_df1 - columns_df2
    columns_only_in_df2 = columns_df2 - columns_df1
    
    if columns_only_in_df1:
        print("Columns only in df1:", columns_only_in_df1)
    else:
        print("No columns only in df1")
    
    if columns_only_in_df2:
        print("Columns only in df2:", columns_only_in_df2)
    else:
        print("No columns only in df2")

# Assuming original_df and features_df are already defined DataFrames
compare_columns(original_df, features_df)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

#features_num_df = num_pipeline.fit_transform(features_num_df)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(features_num_df)
cat_attribs = ["Gender", "Age", "Ethnicity"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

features_prepared = full_pipeline.fit_transform(features_df)
```

```python
# fitting a simple linear regression to predict the median house values

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(features_prepared, targets_df["aveOralF"])
```

```python
# let's see what rmse error the linear regression achieves on the training set

from sklearn.metrics import mean_squared_error

temp_predictions = lin_reg.predict(features_prepared)
lin_mse = mean_squared_error(targets_df["aveOralF"], temp_predictions)
lin_rmse = np.sqrt(lin_mse)
print(round(lin_rmse, 2))
```

```python
# onother useful error is mean absolute error 
# (useful to consider when you suspect outliers in data as it's more robust to outliers than rmse)

from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(targets_df["aveOralF"], temp_predictions)
print(round(lin_mae, 2))
```

```python
# fitting a simple nearest neighbour regression to predict the median house values

from sklearn.neighbors import KNeighborsRegressor
neigh_reg = KNeighborsRegressor(n_neighbors=3)
neigh_reg.fit(features_prepared, targets_df["aveOralF"])
```

```python
# let's see what rmse error the nerest neighbour model achieves on training set;
# we can see here that this model clearly overfits the training set due to a 0 error on training set;
# this is not surprising; why? (see relevant task from the beginning of this lab)

features_predictions = neigh_reg.predict(features_prepared)
neigh_mse = mean_squared_error(targets_df["aveOralF"], features_predictions)
neigh_rmse = np.sqrt(neigh_mse)
neigh_rmse
```

```python
# fitting a simple decision tree with depth 3

from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(features_prepared, targets_df["aveOralF"])
```

```python
# let's see what rmse error the decision tree achieves on training set;
# this is a capable model which can capture non-linear patterns in the data;
# we can see the error on the train set is 0, so it overfits


features_predictions = tree_reg.predict(features_prepared)
tree_mse = mean_squared_error(targets_df["aveOralF"], features_predictions)
tree_rmse = np.sqrt(tree_mse)
print(round(tree_rmse, 2))
```

```python
# the above tree model was fully grown with no limit on its size; let us train one with up to 3 levels
# which is not fully grown so would overfit less

tree_reg = DecisionTreeRegressor(max_depth=4, random_state=42)
tree_reg.fit(features_prepared, targets_df["aveOralF"])
features_predictions = tree_reg.predict(features_prepared)
tree_mse = mean_squared_error(targets_df["aveOralF"], features_predictions)
tree_rmse = np.sqrt(tree_mse)
print(round(tree_rmse, 2))
```

```python
# as seen in class, a random forest is a set of trees working together attempting to achieve a better prediction;
# fitting a simple random forest model consisting of an ensemble of 100 trees (n_estimators) here
# whose predictions are averaged by the random forest model;
# ensemble models, in particular random forests, are powerful as they combine the results produced by several models

from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(features_prepared, targets_df["aveOralF"])
```

```python
# evaluating the random forest model on the training set;

features_predictions = forest_reg.predict(features_prepared)
forest_mse = mean_squared_error(targets_df["aveOralF"], features_predictions)
forest_rmse = np.sqrt(forest_mse)
print(round(forest_rmse, 2))
```
